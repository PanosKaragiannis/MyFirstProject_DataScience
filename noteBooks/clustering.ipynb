```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn import preprocessing
import seaborn as sns
```

```python
# df = pd.read_csv('../processed_data/trans_users_dev.csv')
df = pd.read_csv('../processed_data/trans_users_dev1.csv')
tr1 = pd.read_csv('../data/transactions_1.csv')
tr2 = pd.read_csv('../data/transactions_2.csv')
tr3 = pd.read_csv('../data/transactions_3.csv')
outBount = pd.concat([tr1, tr2, tr3])

df['gg'] = df['number_transactions'] / df['user_active_time'] 
df['gg1'] = df['user_trans_time'] / df['user_active_time'] 
df['avg_amount'] = df['total_amount'] / df['user_active_time'] 

df.head(5)
```

```python
outBount = outBount[ outBount['transactions_state'] == 'COMPLETED' ]
outBount = outBount[ outBount['direction'] == 'OUTBOUND' ]

outBount1 = outBount.groupby(['user_id'], as_index=True)['amount_usd'].agg(['sum', 'count'])
outBount1.rename(columns={'sum':'profit' , 'count':'number_trans_profit'}, inplace=True)
outBount1.head(4)
outBount_n = pd.merge(outBount1, df[['user_id', 'total_amount', 'user_trans_periodicity']], on='user_id')

n_norm = preprocessing.normalize(outBount_n[['number_trans_profit', 'user_trans_periodicity']])
c5 = n_norm[:, [0, 1]] # [[0, 1]] # cluster3
c5
```

```python
N_scaler = MinMaxScaler()
S_scaler = StandardScaler()
c_N = ['total_amount', 'age', 'number_transactions', 'user_active_time', 'avg_amount', 'user_trans_periodicity']
# c2_N = [, 'number_transactions']
# c3_N = ['total_amount']

x = df[c_N].values

N_scaler.fit(x)
x = N_scaler.transform(x)

n_values = preprocessing.normalize(df[c_N])
# if n_values.all() == x.all():
#     print('yes')

# for i in range(len(x)):
#     for j in range(len(x[0])):
#         if x[i, j] != n_values[i, j]:
#             print("fk")
```

```python
cluster1 = [c_N[0], c_N[3]]
cluster2 = [c_N[0], c_N[2]]
cluster3 = [c_N[0], c_N[1]]
cluster4 = [c_N[4], c_N[2]]

c1 = n_values[:, [0, 3]] # [[0, 3]] # cluster1
c2 = n_values[:, [0, 2]] # [[0, 2]] # cluster2
# c2 = c2[c2['total_amount'] < 0.2]
c3 = n_values[:, [0, 1]] # [[0, 1]] # cluster3
c4 = n_values[:, [0, 5]] # [[0, 1]] # cluster3
# c3
```

# Deciding the number of Clusters

```python
silhouette_scores = []
for k in range(2,10):
  kmeans = KMeans(k)
  kmeans.fit(c1)
  preds = kmeans.predict(c1)
  silhouette_scores.append(silhouette_score(c1, preds))

plt.figure(figsize=(10, 6))
ax = plt.subplot(1,1,1)
ax.plot(range(2,10), silhouette_scores)
ax.set_xlabel('k')
ax.set_ylabel('score')
```

# Clustering Based on Amount - Number of Transactions

```python
k = 2

kmeans = KMeans(n_clusters=k, init ='k-means++', max_iter=300, n_init=10, random_state=0)

kmeans.fit(c1)

preds = kmeans.predict(c1)
preds
```

```python
#6 Visualising the clusters
cl1_age = []
cl1_tr = []
cl2_age = []
cl2_tr = []
cl3_age = []
cl3_tr = []
for i in range(len(preds)):
    if preds[i] == 0:
        cl1_age.append(c1[i, 0])
        cl1_tr.append(c1[i, 1])
    else:
        cl2_age.append(c1[i, 0])
        cl2_tr.append(c1[i, 1])
#     else:
#         cl3_age.append(c1.loc[i, 'age'])
#         cl3_tr.append(c1.loc[i, 'number_transactions'])


plt.figure(figsize=(15, 10))
plt.scatter(cl1_age, cl1_tr, s=50, c='green', label ='Cluster 1')
plt.scatter(cl2_age, cl2_tr, s=50, c='blue', label ='Cluster 2')
# plt.scatter(cl3_age, cl3_tr, s=50, c='red', label ='Cluster 3')
# plt.scatter(X[y_kmeans==2, 0], X[y_kmeans==2, 1], s=100, c='green', label ='Cluster 3')
# plt.scatter(X[y_kmeans==3, 0], X[y_kmeans==3, 1], s=100, c='cyan', label ='Cluster 4')
# plt.scatter(X[y_kmeans==4, 0], X[y_kmeans==4, 1], s=100, c='magenta', label ='Cluster 5')
#Plot the centroid. This time we're going to use the cluster centres  #attribute that returns here the coordinates of the centroid.
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label = 'Centroids')
plt.title('Clusters of Users')
plt.xlabel('total amount')
plt.ylabel('user active number')
plt.show()
```
