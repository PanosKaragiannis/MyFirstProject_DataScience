```python
mask = np.zeros_like(tableResult.corr(), dtype=np.bool) 
mask[np.triu_indices_from(mask)] = True 
f, ax = plt.subplots(figsize=(22, 12))
plt.title('Pearson Correlation Matrix',fontsize=25)

sns.heatmap(tableResult.corr(),linewidths=0.25,vmax=1.0,square=True,cmap="GnBu", 
            linecolor='w',annot=True,mask=mask,cbar_kws={"shrink": .75});
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(tableResult.groupby([
        'plan', 'user_settings_crypto_unlocked' ]).volume_transactions.mean().unstack(),annot=True, fmt=".0f")
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(tableResult.groupby([
        'plan', 'user_settings_crypto_unlocked']).number_transactions.mean().unstack(),annot=True, fmt=".0f")
```
